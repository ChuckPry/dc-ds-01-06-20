# Statistical Power

## What is Statistical Power?

The probability that it will reject a false null-hypothesis.

> Power is the likelihood that a study will detect an effect when there is an effect there to be detected.


Recall that we defined $\beta$ as the probability of making a type-II error. 

Therefore: power = 1- $\beta$


> If the effect is bigger than more easier to detect the effect --> more power.

> If the sample size is large than the tests are usually more sensitive to the effect --> more power.

[How do you calculate power?](https://effectsizefaq.com/2010/05/31/how-do-i-calculate-statistical-power/)

```python
from statsmodels.stats.power import TTestIndPower
```

```python
analysis = TTestIndPower()

# parameters for power analysis
effect = 0.7
alpha = .05

power = .9
# perform power analysis
analysis = TTestIndPower()
result = analysis.solve_power(effect, power=power, nobs1= None, ratio=2.0, alpha = alpha)
print('Sample Size: %.3f' % result)
```

```python
import numpy as np
```

```python
from numpy import array
from matplotlib import pyplot
%matplotlib inline
from statsmodels.stats.power import TTestIndPower
# parameters for power analysis
effect_sizes = array([0.2, 0.5, 0.8])
sample_sizes = array(range(5, 100))
alpha_sizes = array([0.08, 0.04,0.025])
# calculate power curves from multiple power analyses
analysis = TTestIndPower()
analysis.plot_power(dep_var='alpha', nobs=sample_sizes, 
                    effect_size = np.array([0.2]), alpha = alpha_sizes)
pyplot.show()
```
